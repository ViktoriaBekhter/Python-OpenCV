
Составить регулярные выражения и с помощью модуля re выделить из приведенного выше текста:
- все номера  телефонов;
- все адреса электронной почты;

# Продвинутые техники поиска подстрок

В следующем тексте с помощью регулярных выражений выделить отдельно доллары, евро, рубли.
```
Компания веселых молодых людей решила хорошо провести время
  У них было:
  2 литра красного вина 12 литров пива 21 $ 12 USD 6100 RUR 35EUR и еще 51 руб. денег
  и пару головок сыра



```




Если использовать традиционный паттерн для цифр **[0-9]+** получим все числа без разделений.

````python
import re

s = '''Компания веселых молодых людей решила хорошо провести время
  У них было:
  2 литра красного вина 12 литров пива 21 $ 12 USD 6100 RUR 35EUR и еще 51 руб. денег   и пару головок сыра'''
pattern = r'[0-9]+'
for item in re.findall(pattern,s):
  print(item)
````

````python
# воспользуемся отдельными шаблонами.
print('доллары')
pattern = r'[0-9]+\s*(?:\$|USD)'
print( *re.findall(pattern,s), sep=',')
print('евро')
pattern = r'[0-9]+\s*EUR'
print( *re.findall(pattern,s), sep=',')

print('рубли')
pattern = r'[0-9]+\s*(?:RUR|руб)'
print( *re.findall(pattern,s), sep=',')
````

 Для рублей \s* произвольное число пробельных символов (их может и не быть). Шаблон (?:RUR|руб) означает или RUR или руб. При этом знак ?: делает наши скоки несохраняющими, иначе их результата был бы отобран вместо нужного нам выражения.
Для доллара используется аналогичная конструкция, (?:\$|USD) , но значок доллара экранирован .


````python
# избавимся и от символов валюты
#Для это применим опережающий отсеивающий просмотр (?= )

print('доллары')
pattern = r'[0-9]+(?=\s*(?:\$|USD))'
print( *re.findall(pattern,s), sep=',')
print('евро')
pattern = r'[0-9]+(?=\s*EUR)'
print( *re.findall(pattern,s), sep=',')
print('рубли')
pattern = r'[0-9]+(?=\s*(?:RUR|руб))'
print( *re.findall(pattern,s), sep=',')
````


Разработать на языке Python функцию, которая с применением регулярных выражений преобразует строку вида


```
N т. M кг.
```


 в число, соответствующее массе в килограммах.


```
 print(mass("1 т. 250 кг."))

```
1250







````python
import re

def mass(input_string):
    # Ищем числа и десятичные дроби в строке
    matches = re.findall(r'\d+\.\d+|\d+', input_string)

    # Проверяем, что найдены два числа
    if len(matches) == 2:
        # Первое число - тонны, второе число - килограммы
        tons = float(matches[0])
        kilograms = float(matches[1])

        # Переводим тонны в килограммы и складываем с килограммами
        total_kilograms = tons * 1000 + kilograms
        return total_kilograms
    else:
        return None

# Пример использования функции
print(mass("1 т. 250 кг."))
````

# Манипулирование подстроками

Есть описание книг. Заменим в выпускающих данных Москву на М.



```
Некрасов Н. Кому на Руси жить хорошо? - Москва: Вече, 2017.
Спиридонов К.Л. Москва моя. - Москва.: Просвещение, 2015.
```




````python
import re

s = '''Некрасов Н. Кому на Руси жить хорошо? - Москва: Вече, 2017.
    Спиридонов К.Л. Москва моя. - Москва.: Просвещение, 2015.'''
print(re.sub('Москва:', 'М:', s))
````

Рассмотрим другой пример. Даны даты в формате: ММ-ДД-ГГГГ, нужно перевести в формат ДД-ММ-ГГГГ

````python
import re

s = '01-08-2021  05-10-2023'

print(re.sub(r'([0-9]+)-([0-9]+)-([0-9]{4})', r'\1-\2-\3',s))
````

Обозначения **\1 \2 \3** это **ссылки** на ранее найденные с помощью шаблона  **группы** . Они заключены в скобки. Теперь мы расставили их в правильном порядке.



Дан текст с разделенными табуляцией колонками. ФИО, должность, телефон.



```
Иванов И.С.			профессор		8(928)321-12-21
Г.В. Сидоров			доцент		+7(928)321-13-22
Карпова-Ли Г.Т.		доцент		8(113)528-13-54
Гераклит А.			ассистент		+7(908)302-15-14, 8(904)505-28-52
Климова М.К.		ст.пр.		8(907)123-12-11
```



Необходимо преобразовать текст так, чтобы:
- инициалы оказались перед фамилией;
- номер телефона начинался не с 8, а с +7


Работу следует выполнить с помощью модуля re языка программирования Python.


````python
import re

def transform_phone(phone):
    if phone.startswith('8'):
        return '+7' + phone[1:]
    return phone

def transform_text(input_text):
    pattern = r'(\S+)\s+(\S+)\s+(\S+)\s+(\S+)'
    transformed_text = []

    for line in input_text.split('\n'):
        match = re.match(pattern, line)
        if match:
            last_name, initials, position, phone = match.groups()
            transformed_phone = transform_phone(phone)
            transformed_line = f'{initials} {last_name}\t{position}\t{transformed_phone}'
            transformed_text.append(transformed_line)

    return '\n'.join(transformed_text)

input_text = """
Иванов И.С.\t\t\tпрофессор\t8(928)321-12-21
Г.В. Сидоров\t\tдоцент\t+7(928)321-13-22
Карпова-Ли Г.Т.\tдоцент\t8(113)528-13-54
Гераклит А.\t\tассистент\t+7(908)302-15-14, 8(904)505-28-52
Климова М.К.\tст.пр.\t8(907)123-12-11
"""

transformed_text = transform_text(input_text)
print(transformed_text)
````
